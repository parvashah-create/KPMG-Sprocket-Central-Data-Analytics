```python
# Load data
tra_df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/kpmg_cleaned_dataset.xlsx',sheet_name="transactions")
cdg_df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/kpmg_cleaned_dataset.xlsx',sheet_name="customer_demographic")
cad_df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/kpmg_cleaned_dataset.xlsx',sheet_name="customer_address")
tra_df["profit"] = tra_df['list_price'] - tra_df['standard_cost']
```
